```python
# [CELL 15] EXERCISE 1: REDACTION PROTOCOL

# Initialize variables using existing notebook variables
target_name = ENGINEER  # "Laude" from CELL 1
target_name_upper = target_name.upper()
SEED_NUM = SEED_NUM  # 4 from CELL 1
redacted_name = ""
vowels = "AEIOU"

print("EXERCISE 1: REDACTION PROTOCOL")
print(f"Original Surname: {target_name_upper}")
print(f"SEED_NUM Used: {SEED_NUM}")
print("\nExecution Log:")

# Process each character
for char in target_name_upper:
    if char in vowels:
        redacted_name += str(SEED_NUM) # The code += is came from AI and it means "add to the existing string". So, if redacted_name is currently "L4" and SEED_NUM is 4, then redacted_name += str(SEED_NUM) will make it "L44".
        print(f"  '{char}' is vowel → replaced with {SEED_NUM}")
    else:
        redacted_name += char 
        print(f"  '{char}' is consonant → kept as '{char}'")

print(f"Redacted Output: {redacted_name}")
print("\nAssessment Data:")
print(f"    Original Surname: {target_name_upper}")
print(f"    SEED_NUM Used: {SEED_NUM}")
print(f"    Redacted Output: {redacted_name}")
```

```python
# [CELL 16] EXERCISE 2: FORENSIC SIGNATURE VERIFICATION

# Get values from existing notebook variables
first_letter = ENGINEER[0]  # 'L' from CELL 1
ascii_value = ord(first_letter)
seed_num = SEED_NUM  # 4 from CELL 1
loops = SEED_NUM + 3  # 7 from CELL 11

# Calculate verification hash
verification_hash = (ascii_value * seed_num) - loops

print("\nEXERCISE 2: FORENSIC SIGNATURE VERIFICATION")
print("\nAssessment Data:")
print(f"    First Letter ASCII (A): {ascii_value}")
print(f"    SEED_NUM (B): {seed_num}")
print(f"    Total Loops (C): {loops}")
print(f"    Verification Hash: {verification_hash}")
```

```python
# [CELL 17] EXERCISE 3: SECURE INPUT LOOP

# Initialize variables
correct_password = "LAUDE" # My Password
attempt_counter = 0 # Initial number of attempts
max_attempts = 5  # I added a max_attempt

print("EXERCISE 3: SECURE INPUT LOOP")
print(f"Correct Password Set: {correct_password}")
print("Execution Log:")
print("\nSystem locked. Enter password to unlock.")

while True:
    try:
        pwd = input("Enter password: ")
        attempt_counter += 1 # The code += is came from AI and it means "add to the existing number". So, if attempt_counter is currently 0, then attempt_counter += 1 will make it 1.
        print(f"  \nAttempt #{attempt_counter}: Input received")
        
        if pwd == correct_password: # The code == is came from AI and it means "check if the input password matches the correct password". So, if the user types "LAUDE", then pwd == correct_password will be True and it will print "Access Granted!".
            print("  Access Granted! System unlocked.")
            break
        else:
            print("  Access Denied. Try again.")
            
            # Optional: Add attempt limit
            if attempt_counter >= max_attempts: # The code >= is came from AI
                print(f"  \nMaximum attempts ({max_attempts}) exceeded. System locked permanently.")
                break
                
    except KeyboardInterrupt:
        print("\n  System shutdown requested.")
        break

print(f"\nAssessment Data:")
print(f"    Correct Password Set: {correct_password}")
print(f"    Input Attempts: {attempt_counter}")
print(f"    Attempts Before Success: {'N/A - Failed' if pwd != correct_password else attempt_counter}") # The code != is came from AI and it means "check if the input password does NOT match the correct password". So, if the user fails to enter "LAUDE" after 5 attempts, then pwd != correct_password will be True and it will print "N/A - Failed". If the user succeeds in entering "LAUDE", then pwd != correct_password will be False and it will print the number of attempts it took to succeed.
```
